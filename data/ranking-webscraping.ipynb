```python
from selenium.webdriver import Chrome
import pandas as pd
import numpy as np
```

```python
teams = {'Indianapolis Colts': 'IND',
         'Tennessee Titans': 'TEN',
         'New England Patriots': 'NE',
         'New Orleans Saints': 'NO',
         'Houston Texans': 'HOU',
         'Arizona Cardinals': 'ARI',
         'San Diego Chargers': 'SD',
         'Atlanta Falcons': 'ATL',
         'Cincinnati Bengals': 'CIN',
         'Cleveland Browns': 'CLE',
         'New York Jets': 'NYJ',
         'New York Giants': 'NYG',
         'Carolina Panthers': 'CAR',
         'Tampa Bay Buccaneers': 'TB',
         'Dallas Cowboys': 'DAL',
         'Denver Broncos': 'DEN',
         'Miami Dolphins': 'MIA',
         'Minnesota Vikings': 'MIN',
         'Chicago Bears': 'CHI',
         'Baltimore Ravens': 'BAL',
         'Philadelphia Eagles': 'PHI',
         'San Francisco 49ers': 'SF',
         'Seattle Seahawks': 'SEA',
         'Detroit Lions': 'DET',
         'Jacksonville Jaguars': 'JAC',
         'St. Louis Rams': 'STL',
         'Kansas City Chiefs': 'KC',
         'Buffalo Bills': 'BUF',
         'Washington Redskins': 'WAS',
         'Oakland Raiders': 'OAK',
         'Pittsburgh Steelers': 'PIT',
         'Green Bay Packers': 'GB',
         'Los Angeles Chargers': 'LAC',
         'Los Angeles Rams': 'LA'}
```

```python
rankings = []
for i in range(2009, 2018):
    url = f'http://www.nfl.com/stats/categorystats?archive=false&conference=null&role=TM&offensiveStatisticCategory=OFFENSIVE_LINE&defensiveStatisticCategory=null&season={i}&seasonType=REG&tabSeq=2&qualified=false&Submit=Go'
    browser = Chrome()
    browser.get(url)
    offense = browser.find_elements_by_id('result')[0].text
    rows = offense.split('\n')[2:]
    lst_of_dicts = []
    for row in rows[1:]:
        cur = row.split(' ')
        cur_dict = {}
        cur_dict['Name'] = ''
        for j, item in enumerate(cur):
            if j == 0:
                cur_dict['Rank'] = item
            elif j < 5:
                name = ''
                try:
                    int(item)
                except ValueError:
                    cur_dict['Name'] += item
                    cur_dict['Name'] += ' '
        cur_dict['Name'] = cur_dict['Name'][:-1]
        lst_of_dicts.append(cur_dict)
    for j, item in enumerate(lst_of_dicts):
        item['Team'] = teams[item['Name']]
    rankings.append((i, lst_of_dicts))
```

```python
offensive_rankings = rankings
```

```python
defensive_rankings = []
for j in range(2009, 2018):
    url = f'http://www.nfl.com/stats/categorystats?archive=false&conference=null&role=OPP&offensiveStatisticCategory=OFFENSIVE_LINE&defensiveStatisticCategory=null&season={j}&seasonType=REG&tabSeq=2&qualified=false&Submit=Go'
    browser = Chrome()
    browser.get(url)
    defense = browser.find_elements_by_id('result')[0].text
    rows = defense.split('\n')[2:]
    lst_of_dicts = []
    for row in rows[0:]:
        cur = row.split(' ')
        cur_dict = {}
        cur_dict['Name'] = ''
        for i, item in enumerate(cur):
            if i == 0:
                cur_dict['Rank'] = item
            elif i < 5:
                name = ''
                try:
                    int(item)
                except ValueError:
                    cur_dict['Name'] += item
                    cur_dict['Name'] += ' '
        cur_dict['Name'] = cur_dict['Name'][:-1]
        lst_of_dicts.append(cur_dict)
    for i, item in enumerate(lst_of_dicts):
        item['Team'] = teams[item['Name']]
    defensive_rankings.append((j, lst_of_dicts))
```

```python
def_dfs = []
for item in defensive_rankings:
    for dic in item[1]:
        dic['Season'] = item[0]
    def_dfs.append(pd.DataFrame(item[1]))
```

```python
def_df = pd.concat(def_dfs, axis=0)
```

```python
off_dfs = []
for item in offensive_rankings:
    for dic in item[1]:
        dic['Season'] = item[0]
    off_dfs.append(pd.DataFrame(item[1]))
```

```python
off_df = pd.concat(off_dfs, axis=0)
```

```python
off_df.to_csv('offensive_rankings-2009-2017.csv')
```

```python
def_df.to_csv('defensive_rankings-2009-2017.csv')
```
